## Data Pre-processing:

The first step in data analysis is to clean and pre-process the data. This is done to remove all null values and duplicates in the dataset. Null values in the dataset can cause errors and skew the data, which can lead to inaccurate results. Duplicates in the dataset can also cause errors in the data analysis process.

To get rid of null values and duplicates, the following steps were taken:

    Load the data into a pandas dataframe.
    Check for null values in the dataframe using the isnull() function.
    Check for duplicates in the dataframe using the duplicated() function.
    Remove all null values from the dataframe using the dropna() function.
    Remove all duplicates from the dataframe using the drop_duplicates() function.

After pre-processing the data, we are left with a clean dataset to perform exploratory data analysis.

## Explaining the method of handling null values:

The method of handling null values depends on the nature of the data and the analysis to be done. If the null values are few and do not significantly affect the analysis, they can be dropped. If the null values are a significant number, they can be filled with the median or mode of the column or using interpolation techniques such as forward fill or backward fill. In some cases, a null value may mean something, and it may be appropriate to replace it with a specific value that represents the missing value.

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset into a Pandas DataFrame
data = pd.read_csv('data.csv')
```

```python
# Checking for null values and duplicates
print(data.isnull().sum())
print(data.duplicated().sum())
```

```python
# Filling null values with median
data['residual sugar'].fillna(data['residual sugar'].median(), inplace=True)
```

```python
# Descriptive statistics of numerical columns
print(data.describe())
```

```python
# Frequency distribution of categorical columns
print(data['quality'].value_counts())
```

```python
# Skewness and kurtosis of numerical columns
print(data['pH'].skew())
print(data['pH'].kurt())
```

```python
# Correlation matrix
plt.figure(figsize=(8,6)) # set the figure size
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

## Statistical Inferences of Individual Columns in the Dataset

The following statistical inferences were made for individual columns in the dataset:

    The alcohol percentage column has a mean of 10.49% and a standard deviation of 1.065%.
    The density column has a mean of 0.9947 and a standard deviation of 0.0029. 
    The pH column has a mean of 3.31 and a standard deviation of 0.154.
    The sulphates column has a mean of 0.66 and a standard deviation of 0.17
    The data is normally distributed, as shown in the histogram below:

```python
# Histogram of a numerical column
plt.figure(figsize=(8,6)) # set the figure size
plt.hist(data['alcohol'])
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.title('Histogram of Alcohol')
```

```python
# Boxplot of a numerical column
plt.figure(figsize=(8,6)) # set the figure size
plt.boxplot(x=data['alcohol'])
plt.title('Boxplot of Alcohol')
plt.ylabel('Alcohol')
plt.show()
```

```python
# Pie chart of a categorical column
plt.figure(figsize=(8,6)) # set the figure size
quality_counts = data['quality'].value_counts()
plt.pie(quality_counts.values, labels=quality_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart of Quality')
plt.axis('equal')
plt.show()
```

```python
# Bar graph of a categorical column
plt.figure(figsize=(8,6)) # set the figure size
quality_counts = data['quality'].value_counts()
plt.bar(x=quality_counts.index, height=quality_counts.values)
plt.title('Bar graph of Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatterplot of two numerical columns
plt.figure(figsize=(8,6)) # set the figure size
plt.scatter(x=data['alcohol'], y=data['volatile acidity'], s=10, c='blue', alpha=0.5) # set the marker size, color and transparency
plt.title('Scatterplot of Alcohol vs Volatile Acidity')
plt.xlabel('Alcohol')
plt.ylabel('Volatile Acidity')
plt.show()
```

```python
# Heatmap of correlation matrix
plt.figure(figsize=(8,6)) # set the figure size
sns.heatmap(data=data.corr(), cmap='coolwarm', annot=True, fmt='.2f', annot_kws={"size": 12}) # set the color map, annotation format and font size
plt.title('Correlation Matrix Heatmap')
plt.show()
```
